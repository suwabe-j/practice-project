```python
# CSV形式でデータフレームとして読み込んで、正しく読めたか確認するために上位３つを表示する
import pandas as pd
ex4_csv = pd.read_csv("ex4.csv")
print(ex4_csv.head(3))

# 欠損値の確認
print(ex4_csv.isnull().any(axis=0))

# 特徴量である性別（sex）での割合を確認して、不均一かどうか確認する（引数に１を入れて、割合を出す）
print(ex4_csv["sex"].value_counts(1)*100)

# classごとにscoreの平均値を出す
print(ex4_csv.groupby("class")["score"].mean())

# classごとに性別でどう平均値が違うのかを出す
print(pd.pivot_table(ex4_csv, index="class", columns="sex", values="score", aggfunc="mean"))

# dept_idをダミー数変化させる
dept_id_dummy = pd.get_dummies(ex4_csv["dept_id"], drop_first=True, dtype=int)
print(dept_id_dummy)

# ダミー数変化させたものを横方向に元データと連結させる
ex4_concat_dummy = pd.concat([ex4_csv,dept_id_dummy], axis=1)
print(ex4_concat_dummy)

# もとのdept_idを列削除
ex4_end_csv = ex4_concat_dummy.drop("dept_id", axis=1)
print(ex4_end_csv)

```
